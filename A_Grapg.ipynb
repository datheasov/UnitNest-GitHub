```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# === 1) Define your table data as a DataFrame ===
data = {
    "Epoch": [50, 50, 75, 75, 100, 100],
    "Model": ["CNN", "ResNet", "CNN", "ResNet", "CNN", "ResNet"],
    "Precision": [87.10, 75.86, 97.22, 59.52, 100.00, 71.43],
    "Recall":    [65.85, 53.66, 85.37, 85.29, 82.39, 52.39],
    "F1-Score":  [75.00, 62.86, 90.91, 67.92, 90.67, 60.95],
    "Accuracy":  [78.05, 68.29, 91.46, 69.46, 91.46, 70.00]
}
df = pd.DataFrame(data)

# === 2) Setup plotting details ===
metrics = ["Precision", "Recall", "F1-Score", "Accuracy"]  # columns to plot
epochs = [50, 75, 100]                                     # each subplot
colors = ['dodgerblue', 'skyblue', 'deepskyblue', 'lightskyblue']
# ^ one color per metric

# Create 3 subplots side-by-side (one for each epoch), sharing the y-axis
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True, gridspec_kw={'wspace': 0.05})

# === 3) Generate bar charts for each epoch ===
for idx, epoch in enumerate(epochs):
    # Filter rows for this epoch
    epoch_df = df[df["Epoch"] == epoch].copy()
    # epoch_df now has 2 rows: one for CNN, one for ResNet
    
    # Extract just the numeric values for these metrics, shape => (4 metrics, 2 models)
    values = epoch_df[metrics].values.T
    
    # x positions for the 2 models (CNN, ResNet)
    x = np.arange(len(epoch_df["Model"].unique()))  # -> array([0, 1])
    bar_width = 0.2
    
    ax = axes[idx]
    
    # Loop over each metric to plot
    for i, metric in enumerate(metrics):
        # values[i] => array of length 2 (CNN and ResNet) for this metric
        bars = ax.bar(
            x + i * bar_width,         # shift each metric group
            values[i], 
            width=bar_width,
            color=colors[i],
            label=metric if idx == 0 else None  # Add legend labels only on first subplot
        )
        # Annotate each bar
        for bar in bars:
            height = bar.get_height()
            ax.annotate(
                f"{height:.2f}",               # 2 decimals
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3), 
                textcoords="offset points",
                ha='center', va='bottom', 
                fontsize=9,
                rotation=90
            )
    
    # Subplot title and axes
    ax.set_title(f"{epoch} Epochs", fontsize=11)
    ax.set_xticks(x + bar_width * (len(metrics) / 2 - 0.5))
    ax.set_xticklabels(["CNN", "ResNet"])
    ax.set_ylim(0, 110)  # Enough headroom for 100% + label offset
    ax.set_xlabel("Model")

# Y-axis label on the leftmost subplot
axes[0].set_ylabel("Percentage (%)")

# === 4) Single legend centered below all subplots ===
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', bbox_to_anchor=(0.5, -0.08), ncol=len(metrics), fontsize=10)

# Adjust layout to remove extra whitespace
plt.subplots_adjust(left=0.07, right=0.95, top=0.88, bottom=0.25)
# Save as SVG
svg_path = "comparison_classification_all_epoch.svg"
plt.savefig(svg_path, format='svg')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = ["CNN", "ResNet50"]
correct = [56, 42]
incorrect = [4, 18]

x = np.arange(len(models))  # [0, 1]
width = 0.35                # width of each bar

# Create the figure and axes
fig, ax = plt.subplots(figsize=(6, 6))

# Plot the bars
rects_correct = ax.bar(x - width/2, correct, width, label='Correct', color='blue')
rects_incorrect = ax.bar(x + width/2, incorrect, width, label='Incorrect', color='lightskyblue')

# Add labels, title, legend
ax.set_ylabel("Number of Predictions")
ax.set_title("Classification Performance (60 Test Images)")
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()

# Function to attach a text label above each bar
def add_bar_labels(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(
            f"{height}",
            xy=(rect.get_x() + rect.get_width() / 2, height),
            xytext=(0, 3),   # vertical offset
            textcoords="offset points",
            ha='center', va='bottom'
        )

# Label both sets of bars
add_bar_labels(rects_correct)
add_bar_labels(rects_incorrect)

# Adjust y-limit to make sure labels fit
ax.set_ylim(0, 60)

plt.tight_layout()
plt.show()
```
